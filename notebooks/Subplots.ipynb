```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()

for i in range(1,7):
    plt.subplot(3,2,i)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])
    
    plt.tight_layout()
    ;
```

```python
plt.figure(figsize=(10, 10))

x = np.arange(1,21)
a, b = 2, 0.6

for i in range(1,17):
    y_i = a + b*x + np.random.normal(0,1,20)
    plt.subplot(4,4,i)
    plt.plot(x,y_i)
    plt.title('Series {}'.format(i))
    plt.xticks([0,5,10,15,20],['2010','2011','2012','2013','2014'])
    plt.tight_layout()
```
