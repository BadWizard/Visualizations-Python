```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

## Basic line plots

```python
x = np.linspace(0, 20)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)
```

```python
plt.figure()

plt.plot(x,
        y1,
        color='black',
        linestyle = '-',
        linewidth = 2, 
        marker = 's',
        markersize=6,
        label='y1')

plt.plot(x,
        y2,
        color = 'gray',
        linestyle='--',
        linewidth=2,
        marker='^',
         markersize=6,
        label='y2' )
plt.legend()
;
```

## Basic bar plots

```python
years = np.arange(2012,2015)
values = [2,5,9]
```

```python
plt.figure()

plt.bar(years, values)
plt.xticks(years,[str(year) for year in years ])
;
```

```python
years = np.arange(2012, 2015)
category1_values = [2,5,9]
category2_values = [7,6,3]

plt.figure()

plt.bar(years - 0.2,
       category1_values,
       color='blue',
       edgecolor='none',
       width=0.4,
       align='center',
       label='y1')

plt.bar(years + 0.2,
       category2_values,
       color='orange',
       edgecolor='none',
       width=0.4,
       align='center',
       label='y2')

plt.xticks(years,[str(year) for year in years ])

plt.legend()
```

## Horizontal bar charts
### for comparing categories

```python
categories = ['A', 'B','C', 'D', 'F']
values = [7, 12, 4, 2, 9]

plt.figure()

plt.barh(np.arange(len(categories)), values)

plt.yticks(np.arange(len(categories)), ['category {}'.format(cat) for cat in categories]);
```

```python
categories = ['A', 'B','C', 'D', 'F']
values = [7, 12, 4, 2, 9]

plt.figure()

plt.barh(-np.arange(len(categories)),
         values,
        color='blue',
        edgecolor='none',
        height=0.6,
        align='center')

plt.yticks(-np.arange(len(categories)), ['category {}'.format(cat) for cat in categories]);
```

```python
categories = ['A', 'B','C', 'D', 'F']
values = [7, 12, 4, 2, 9]

category_values = pd.DataFrame({'categories':categories,
                               'values':values})
category_values.sort(columns='values', inplace=True, ascending=False)

categories = category_values.categories
values = category_values.values

plt.figure()

plt.barh(-np.arange(len(categories)))
```
